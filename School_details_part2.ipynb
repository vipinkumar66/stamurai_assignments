```python
#importing importnat libraries
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import csv
import re
```

```python
#requesting page
page= requests.get("https://www.pwcs.edu/academics___programs/index#")
soup= bs(page.content, "lxml")
selecting_school_links= soup.find("ul", class_= "schoolList")
school_list= selecting_school_links.find_all("li")

#getting the school links
school_links=[]
for listt in school_list:
    if listt.a == None:
        pass
    else:
        school_links.append(listt.a['href'])

school_links
```

```python
#iterating over each school and getting the staff directory link
staff_dir= []
def school_web(link):
    page= requests.get(link)
    soup= bs(page.content, "lxml")
    try:
        menu_bar= soup.find("ul", class_="nav-menu")
        lists= menu_bar.find_all("li", class_="nav-item m_item level0 parent")
        #return lists
        for lis in lists:
            a_tags = lis.find_all("a", class_ ="current")
            for a_tag in a_tags:
                dir_link=""
                if a_tag.text == "Staff Directory" :
                    dir_link = (a_tag['href'])
                    staff_directory_link= link + "/" + dir_link +".json"
                    staff_dir.append(staff_directory_link)
                else:
                    pass
    except Exception as e:
        pass
        
for linkk in school_links:                
    school_web(linkk)
```

```python
#getting the teacher details for each school

def teacher_details(link):
    response = requests.get(link)
    if response.ok:
        data = response.json()["faculty"]
        school_name= (link.split(".")[0])
        schoolname= school_name.split("//")[1].upper()
        #['School Name']= schoolname.upper()
        for x in data:
            x.update({'school name':schoolname})
                     
        with open('school_details_2.csv', 'a') as testfile:
            fieldnames= ['fullName', 'title', 'email', 'school name']
            writer = csv.DictWriter(testfile, fieldnames=fieldnames, extrasaction='ignore')
            writer.writeheader()
            writer.writerows(data)
        
for linkk in staff_dir:   
    teacher_details(linkk)  
```
